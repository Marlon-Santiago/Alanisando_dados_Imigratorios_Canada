```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('imigrantes_canada.csv')
dados
```

```python
dados.País
```

```python
dados.info()
```

```python
dados.set_index('País', inplace = True)
```

```python
dados.head(5)
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
brasil = dados.loc['Brasil', anos]
brasil
```

```python
brasil_dic = {'Ano' : brasil.index.tolist(), 'Imigrantes': brasil.values.tolist()}
```

```python
brasil_dados = pd.DataFrame(brasil_dic)
brasil_dados
```

```python
plt.title('Teste de Titulo')
plt.xlabel('Anos')
plt.ylabel('Quantidade de Imigrantes')
plt.plot(brasil_dados['Ano'], brasil_dados['Imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show
```
