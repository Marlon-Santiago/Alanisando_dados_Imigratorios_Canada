```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('imigrantes_canada.csv')
dados
```

```python
dados.País
```

```python
dados.info()
```

```python
dados.set_index('País', inplace = True)
```

```python
dados.head(5)
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
brasil = dados.loc['Brasil', anos]
brasil
```

```python
brasil_dic = {'Ano' : brasil.index.tolist(), 'Imigrantes': brasil.values.tolist()}
```

```python
brasil_dados = pd.DataFrame(brasil_dic)
brasil_dados
```

```python
plt.figure(figsize = (10, 5))
plt.title('Imigrantes do Brasil para no Canada de 1980 a 2013')
plt.xlabel('Anos')
plt.ylabel('Quantidade de Imigrantes')
plt.plot(brasil_dados['Ano'], brasil_dados['Imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(brasil_dados.Ano, brasil_dados.Imigrantes)
ax.set_title('Imigrantes do Brasil para Canada no periodo de \n 1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Quantidade de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(3))
ax.grid()
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize = (15, 7))
axs[0].plot(brasil_dados.Ano, brasil_dados.Imigrantes)
axs[0].set_title('Imigrantes do Brasil para Canada no periodo de \n 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Quantidade imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()


axs[1].boxplot(brasil_dados.Imigrantes)
axs[1].set_title('Imigrantes do Brasil para Canada no periodo de \n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Numeros de imigrantes')
axs[1].grid()

plt.show()
```
