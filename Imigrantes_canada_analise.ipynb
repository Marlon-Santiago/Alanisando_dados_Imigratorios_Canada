```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dados = pd.read_csv('imigrantes_canada.csv')
dados
```

```python
dados.País
```

```python
dados.info()
```

```python
dados.set_index('País', inplace = True)
```

```python
dados.head(5)
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
brasil = dados.loc['Brasil', anos]
brasil
```

```python
brasil_dic = {'Ano' : brasil.index.tolist(), 'Imigrantes': brasil.values.tolist()}
```

```python
brasil_dados = pd.DataFrame(brasil_dic)
brasil_dados
```

```python
plt.figure(figsize = (10, 5))
plt.title('Imigrantes do Brasil para no Canada de 1980 a 2013')
plt.xlabel('Anos')
plt.ylabel('Quantidade de Imigrantes')
plt.plot(brasil_dados['Ano'], brasil_dados['Imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(brasil_dados.Ano, brasil_dados.Imigrantes)
ax.set_title('Imigrantes do Brasil para Canada no periodo de \n 1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Quantidade de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(3))
ax.grid()
plt.show()
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
ax.barh(america_do_sul.index, america_do_sul.Total, color = cores)
ax.set_title('Imigração da América do Sul para o Canadá 1980 a 2013', loc='left', fontsize = 18)
ax.set_ylabel('')
ax.set_xlabel('Quantidade de imigrantes', fontsize = 14)
ax.yaxis.set_tick_params(labelsize = 12)
ax.xaxis.set_tick_params(labelsize = 12)

plt.show()
```

```python
america_do_sul_sorted = america_do_sul.sort_values('Total', ascending = True)
```

```python
cor = []

for pais in america_do_sul_sorted.index:
    
    if pais == 'Brasil':
        cor.append('green')

    else:
        cor.append('black')
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
ax.barh(america_do_sul_sorted.index, america_do_sul_sorted['Total'], color = cor)
ax.set_title('Brasil é o 4º pais com mais imigrantes para o Cánada da america do sul de 1980 a 2013', loc='left', fontsize = 16)
ax.set_ylabel('')
ax.set_xlabel('Quantidade de imigrantes', fontsize = 14)
ax.yaxis.set_tick_params(labelsize = 12)
ax.xaxis.set_tick_params(labelsize = 12)

for i, v  in enumerate(america_do_sul_sorted.Total):
    ax.text(v + 20, i, str(v), color = 'black', fontsize = 12, ha = 'left', va = 'center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis = 'both', which = 'both', length = 0)

plt.show()
```

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize = (10,5))
    ax = sns.barplot(data = top_10, x = 'Total', y =top_10.index, orient = 'h', palette = palette)
    ax.set_title('Países com maior iimigração para o canada de 1980 a 2013', loc = 'left', fontsize = 16)
    ax.set_xlabel('Quantidade de imigrantes', fontsize = 14)
    ax.set_ylabel('')


    plt.show()
```
